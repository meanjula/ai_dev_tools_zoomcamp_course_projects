```python
import requests
from zipfile import ZipFile
from io import BytesIO
```

```python
url = "https://github.com/jlowin/fastmcp/archive/refs/heads/main.zip"
data = requests.get(url)
```

```python
zf = ZipFile(BytesIO(data.content))
file_dict = {}
for info in zf.infolist():
    filename = info.filename
    if not filename.split(".")[-1] in ["md", "mdx"]:
        continue
    with zf.open(filename) as file:
        content = file.read().decode("utf-8", errors="ignore")
        updated_file_name = '/'.join(filename.split('/')[1:])
        file_dict[updated_file_name] = content
```

```python
folders = [f.split('/')[0] for f in file_dict.keys()]
```

```python
mdFiles = file_dict
print(len(mdFiles))
print(mdFiles.keys())
print("Folders in the repository:",set(folders))
for md in mdFiles:
    print("Markdown Files in the repository:",md)
```
